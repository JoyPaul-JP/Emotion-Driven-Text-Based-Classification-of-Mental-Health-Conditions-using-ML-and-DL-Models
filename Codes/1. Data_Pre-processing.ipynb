```python
import pandas as pd
import nltk
from nltk.corpus import words
from nltk.tokenize import word_tokenize
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer, WordNetLemmatizer
```

```python
# Download necessary NLTK data
nltk.download('words')
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('stopwords', quiet=True)
nltk.download('omw-1.4')  # Optional for WordNet Lemmatizer
```

```python
# Initialize objects for stemming and lemmatization
stemmer = SnowballStemmer(language='english')
lemmatizer = WordNetLemmatizer()
```

```python
# Load English words set
english_words = set(words.words())
```

```python
# Function to filter non-English words for Labels 0 and 2
def filter_non_english_basic(text):
    return ' '.join([word for word in text.split() if word.lower() in english_words])
```

```python
# Extend the list with common stemmed and lemmatized forms
extended_english_words = set()

for word in english_words:
    # Add the original word
    extended_english_words.add(word)
    # Add the stemmed version of the word
    extended_english_words.add(stemmer.stem(word))
    # Add the lemmatized version of the word
    extended_english_words.add(lemmatizer.lemmatize(word))

# Function to filter non-English words for Label 1
def filter_non_english_extended(text):
    return ' '.join([word for word in text.split() if word.lower() in extended_english_words])
```

```python
# Function to apply the correct filter based on label
def conditional_filter(text, label):
    if label == 1:
        return filter_non_english_extended(text)
    else:
        return filter_non_english_basic(text)
```

```python
# Apply the filter conditionally
df['filtered_text_eng'] = df.apply(lambda row: conditional_filter(row['text'], row['Label']), axis=1)
```

```python
# Function for text preprocessing
def preprocess_text(text):

    # Remove URL
    text = re.sub(r'((www\.[^\s]+)|(https?://[^\s]+))', '', text)
    text = re.sub(r'#([^\s]+)', '', text)

    # Remove all the special characters
    text = re.sub(r'\W', ' ', text)

    # Remove all single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)

    # Remove single characters from the start
    text = re.sub(r'\^[a-zA-Z]\s+', ' ', text)

    # Remove numeric
    text = re.sub(r'\d+', '', text)

    # Substituting multiple spaces with single space
    text = re.sub(r'\s+', ' ', text, flags=re.I)

    # Convert to Lowercase
    text = text.lower()

    # Remove whitespace at begin and end string
    text = text.strip()

    # Tokenize the text
    words = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Stemming
    words = [stemmer.stem(word) for word in words]

    # Lemmatization
    words = [lemmatizer.lemmatize(word) for word in words]

    # Join the words back into a string
    processed_text = ' '.join(words)

    return words, processed_text
```

```python
# Apply preprocessing to the 'Text' column
df['tokenized_text'], df['Preprocessed_Text'] = zip(*df['filtered_text_eng'].apply(preprocess_text))
df.head(20)
```
